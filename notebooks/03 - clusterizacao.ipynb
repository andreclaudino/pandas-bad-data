Visualizar ddos de múltiplas feições é complicado, temos apenas duas ou três dimensões, mas muitas vezes temos mais feições que isso. O que podemos fazer é uma redução de dimensionalidade.

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
display(df['familia aceita'].unique())
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer(neg_label=0, pos_label=1, sparse_output=False)
```

```python
transformed = lb.fit_transform(df['familia aceita'].values)
familia_aceita = pd.DataFrame(transformed, columns=lb.classes_).set_index(df.index.values)
familia_aceita
```

```python
# De um modo geral a codificação pode ser feita pela função seguinte
def encode_one_hot(df, column_name):
    transformed = lb.fit_transform(df[column_name].values)
    return pd.DataFrame(transformed, columns=lb.classes_).set_index(df.index.values)
```

```python
municipios = encode_one_hot(df, 'município')
municipios
```

```python
estados = encode_one_hot(df, 'estado')
estados
```
